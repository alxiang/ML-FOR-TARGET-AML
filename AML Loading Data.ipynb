## Load Data

### Load Gene Expression Data

```python
PATH = "AML-DATA/"

# Load in gene expression matrix
masterGen = pd.read_table(PATH+"TARGET_AML_GE_level3.txt", low_memory=False)
masterGen.index = masterGen["ID"]
masterGen.drop(["ID", "Gene Symbol"], axis = 1, inplace = True)

masterGen.head()
```

### Load Clinical Data

```python
# Load in TARGET-AML Discovery Clinical Matrix
discoveryClin = pd.read_excel(PATH+"TARGET_AML_Discovery_ClinicalData_20170525.xlsx")
discoveryClin.index = discoveryClin["TARGET USI"]
discoveryClin.drop("TARGET USI", axis = 1, inplace = True)
discoveryClin = discoveryClin.T

discoveryClin.head()
```

```python
# Load in TARGET-AML Validation Clinical Matrix

validationClin = pd.read_excel(PATH+"TARGET_AML_Validation_ClinicalData_20170525.xlsx")
validationClin.index = validationClin["TARGET USI"]
validationClin.drop("TARGET USI", axis = 1, inplace = True)
validationClin = validationClin.T

validationClin.head()
```

## Export Loaded Datasets


```python
masterGen.to_pickle(PATH+"PD-TARGET-AML_GE)
discoveryClin.to_pickle(PATH+"PD-TARGET-AML-Discovery-ClinicalData")
validationClin.to_pickle(PATH+"PD-TARGET-AML-Validation-ClinicalData")
```
