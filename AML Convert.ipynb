```python
PATH = "AML-DATA/"

X = pd.read_pickle(PATH + "PROCESSED-TARGET-AML-X.pkl")
```

```python
probeIDS = np.array(X.columns)

# File Batches for Affymetrix Exon/Gene Expression Query (takes max 10000 IDs), (Transcript Cluster IDs, Human ST 1.0)
np.savetxt("probeclusterbatches/probeids0.txt", probeIDS[:10000], '%5.0f')
np.savetxt("probeclusterbatches/probeids1.txt", probeIDS[10000:20000], '%5.0f')
np.savetxt("probeclusterbatches/probeids2.txt", probeIDS[20000:30000], '%5.0f')
np.savetxt("probeclusterbatches/probeids3.txt", probeIDS[30000:], '%5.0f')
```

```python
'''
Use Affymetrix Website for processing in this step
'''
```

```python
PATH2 = "exportedgenebatches/"

batch0 = pd.read_csv(PATH2+"export0.tsv", sep='\t')
batch1 = pd.read_csv(PATH2+"export1.tsv", sep='\t')
batch2 = pd.read_csv(PATH2+"export2.tsv", sep='\t')
batch3 = pd.read_csv(PATH2+"export3.tsv", sep='\t')
```

```python
fullBatch = pd.concat([batch0, batch1, batch2, batch3])
```

```python
print(fullBatch.shape, X.shape)
```

```python
fullBatch.index = fullBatch["Transcript Cluster ID"]
```

```python
fullBatch.drop("Transcript Cluster ID", inplace = True, axis = 1)
fullBatch = fullBatch.T
fullBatch.head()
```

```python
X = X.astype("object")
fullBatch = fullBatch.astype("object")
X_mod = pd.concat([fullBatch, X], axis = 0, join = "outer", sort=True)
```

```python
X_mod = X_mod.T
X_mod.shape
```

```python
X_mod = X_mod[X_mod["Gene Symbol"] != "---"]
```

```python
print(X_mod.shape)
X_mod = X_mod.T
X_mod.head()
```

```python
X_mod.columns = X_mod.iloc[1]
X_mod.drop(["Gene Title", "Entrez Gene ID", "Cytoband", "GO Description", "Transcript Classification", "Gene Symbol"],
           axis = 0, inplace = True)
```

```python
# Some genes are now duplicated (had same cluster ID mapping)

X_mod = X_mod.loc[:,~X_mod.columns.duplicated()]
```

```python
X_mod.shape
```

```python
X_mod.to_csv(PATH+"X_mod.csv")
```
